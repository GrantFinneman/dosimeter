```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
import os
```

```python
files = sorted(glob("./dicoms/6MV_results/longitude/*.csv"))
for file in files:
    plt.figure()
    df = pd.read_csv(file)
    plt.plot(df['Distance_(mm)'], df['Gray_Value'], color="indigo")
    plt.title(os.path.basename(file))
    plt.ylabel("Gray Value")
    plt.xlabel("Depth (mm)")
    plt.savefig('./../figures/dicom_plots/' + os.path.basename(file).rstrip('csv') + "svg", format="svg")
files = sorted(glob("./dicoms/10MV_results/longitude/*.csv"))
for file in files:
    plt.figure()
    df = pd.read_csv(file)
    plt.plot(df['Distance_(mm)'], df['Gray_Value'], color="indigo")
    plt.title(os.path.basename(file))
    plt.ylabel("Gray Value")
    plt.xlabel("Depth (mm)")
    plt.savefig('./../figures/dicom_plots/' + os.path.basename(file).rstrip('csv') + "svg", format="svg")
files = sorted(glob("./dicoms/6MV_results/lattitude/*.csv"))
for file in files:
    plt.figure()
    df = pd.read_csv(file)
    plt.plot(df['Distance_(mm)'], df['Gray_Value'], color="indigo")
    plt.title(os.path.basename(file))
    plt.ylabel("Gray Value")
    plt.xlabel("Depth (mm)")
    plt.savefig('./../figures/dicom_plots/' + os.path.basename(file).rstrip('csv') + "svg", format="svg")
files = sorted(glob("./dicoms/10MV_results/lattitude/*.csv"))
for file in files:
    plt.figure()
    df = pd.read_csv(file)
    plt.plot(df['Distance_(mm)'], df['Gray_Value'], color="indigo")
    plt.title(os.path.basename(file))
    plt.ylabel("Gray Value")
    plt.xlabel("Depth (mm)")
    plt.savefig('./../figures/dicom_plots/' + os.path.basename(file).rstrip('csv') + "svg", format="svg")
```
